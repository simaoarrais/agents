```python
# Print the key prefixes to help with any debugging
OLLAMA_HOST = os.getenv("OLLAMA_HOST", "localhost")
OLLAMA_PORT = os.getenv("OLLAMA_PORT", "11434")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY", "ollama")
ANTHROPIC_API_KEY = os.getenv("ANTHROPIC_API_KEY", "ollama")

if OLLAMA_HOST:
   print(f"OLLAMA_HOST exists and begins with: {OLLAMA_HOST[:3]}")
else:
   print("OLLAMA_HOST not set.")

if OLLAMA_PORT:
   print(f"OLLAMA_PORT exists and begins with: {OLLAMA_PORT[:3]}")
else:
   print("OLLAMA_PORT not set.")
   
if OPENAI_API_KEY:
	print(f"OpenAI API Key exists and begins with: {OPENAI_API_KEY[:3]}")
else:
   print("OpenAI API Key not set.")
    
if ANTHROPIC_API_KEY:
   print(f"Anthropic API Key exists and begins with: {ANTHROPIC_API_KEY[:3]}")
else:
   print("Anthropic API Key not set.")
```

```python
# Now we will create a challenging question to ask the LLMs
request = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. "
request += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": request}]
print(messages)
```

## OpenAI API - Ollama

```python
# Get the question from the local Ollama LLM using the OpenAI API client
ollama = OpenAI(
   base_url=f"http://{OLLAMA_HOST}:{OLLAMA_PORT}/v1", 
   api_key=OPENAI_API_KEY # required but ignored
)
model_name = "llama3.2:1b"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content
print(answer)
```

```python
# Now we will ask the generated question to the LLMs
messages = [{"role": "user", "content": answer}]
response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content
print(answer)
```

```python
# Create a challenging question to ask the LLMs
request = "Please come up with a challenging, nuanced question that I can ask a number of LLMs to evaluate their intelligence. "
request += "Answer only with the question, no explanation."
messages = [{"role": "user", "content": request}]
print(messages)

# Get the question from the local Ollama LLM using the OpenAI API client
ollama = OpenAI(
   base_url=f"http://{OLLAMA_HOST}:{OLLAMA_PORT}/v1", 
   api_key=OPENAI_API_KEY # required but ignored
)
model_name = "llama3.2:1b"

response = ollama.chat.completions.create(model=model_name, messages=messages)
question = response.choices[0].message.content
display(Markdown(question))
```

```python
# Ask generated question to Llama3.2:1b
answer_map = {}
messages = [{"role": "user", "content": question}]
response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content
answer_map[model_name] = answer
display(Markdown(answer))
```

```python
# Ask generated question to Deepseek-r1:1.5b
ollama = OpenAI(
   base_url=f"http://{OLLAMA_HOST}:{OLLAMA_PORT}/v1", 
   api_key=OPENAI_API_KEY # required but ignored
)
model_name = "deepseek-r1:1.5b"

response = ollama.chat.completions.create(model=model_name, messages=messages)
answer = response.choices[0].message.content
answer_map[model_name] = answer
display(Markdown(answer))
```

```python
# Print all answers together for comparison
together = ""
for index, answer in enumerate(answer_map.values()):
	together += f"# Response from competitor {index+1}\n\n"
	together += answer + "\n\n"
print(together)
```

```python
judge = f"""You are judging a competition between {len(answer_map)} competitors.
Each model has been given this question:

{question}

Your job is to evaluate each response for clarity and strength of argument, and rank them in order of best to worst.
Respond with JSON, and only JSON, with the following format:
{{"results": ["best competitor number", "second best competitor number", "third best competitor number", ...]}}

Here are the responses from each competitor:

{together}

Now respond with the JSON with the ranked order of the competitors, nothing else. Do not include markdown formatting or code blocks."""
```

```python
judge_messages = [{"role": "user", "content": judge}]

openai = OpenAI(
   base_url=f"http://{OLLAMA_HOST}:{OLLAMA_PORT}/v1", 
   api_key=OPENAI_API_KEY # required but ignored
)
model_name = "llama3.2:1b"

response = openai.chat.completions.create(
	model=model_name,
	messages=judge_messages,
)
results = response.choices[0].message.content
print(results)
```

```python
results_dict = json.loads(results)
ranks = results_dict["results"]
models = list(answer_map.keys())
for index, result in enumerate(ranks):
    num = int(''.join(filter(str.isdigit, result)))
    competitor = models[num - 1]
    print(f"Rank {index+1}: {competitor}")
```
